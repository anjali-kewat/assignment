```python
#1
class bankaccount():

    def  __init__(self,account_number ,account_hoder,balance =0):
        self.account_number = account_number
        self.account_holder = account_hoder
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance +=amount
            return f"Deposit sucessful. New balance: {self.balance}"
        else:
            return f"Invalid deposit amount .Pleaze enter positive amount"
        
    def withdraw(self,amount):
        if 0< amount <= self.balance:
            self.balance -= amount
            return f"Withdraw sucessful.New balance : {self.balance} "
        else:
            return f"Invalid withdrawl a(mount or insufficient funds "
        
account1 = bankaccount("123456789","devi nishad",1000)
print(account1.deposit(500))
print(account1.withdraw(100))

#2.
class employee():
    
    def __init__(self , employee_id, name , salary):
        self.employee_id = employee_id
        self.name = name 
        self.salary = salary

    def calculated_yearly_bonus(self):
        bonus_percentage = 0.10
        yearly_bonus = self.salary*bonus_percentage
        return yearly_bonus
    
    def display_emplyee_details(self):
        details = f"Employee Id : {self.employee_id}\n Name : {self.name}\nSalary: {self.salary}\n "
        return details 
    

emplyee1 = employee("E001","devi nishad",60000)
bonus = emplyee1.calculated_yearly_bonus()
details = emplyee1.display_emplyee_details()

print(details)
print(f"Yearly Bonus : rs.{bonus}")

class vehicleRentalSystem():

    def __init__(self):
        self.availaible_vehicle = set()
        self.rented_vehicle= set()

    def add_vehicle(self,vehicle_id ):
        self.availaible_vehicle.add(vehicle_id)

    def rent_vehicle(self,vehicle_id):
        if vehicle_id in self.availaible_vehicle:
            self.availaible_vehicle.remove(vehicle_id)
            self.rented_vehicle.add(vehicle_id)
            return f"Vehicle {vehicle_id} rented sucessfully."
        else:
            return f"Vehicle {vehicle_id} is not availaible for rent ."

    def return_vehicle(self,vehicle_id):
        if vehicle_id in self.rented_vehicle:
            self.rented_vehicle.remove(vehicle_id)
            self.availaible_vehicle.add(vehicle_id)
            return f"vehicle {vehicle_id}  returned sucessfully"
        else:
            return f"Vehicle {vehicle_id} was not returned sucessfully"

    def display_availaible_vehicles(self):
        return f"Availaible vehicles: {', '.join(self.availaible_vehicle) } " if self.availaible_vehicle else "No vehicles availaible"
    
rental_system = vehicleRentalSystem()
rental_system.add_vehicle("Car001")
rental_system.add_vehicle("Bike001")
rental_system.add_vehicle("scooter001")

print(rental_system.display_availaible_vehicles())

print(rental_system.rent_vehicle("Car001"))
print(rental_system.display_availaible_vehicles())

print(rental_system.return_vehicle("Car001"))
print(rental_system.display_availaible_vehicles())

#4.

class Book:

    def __init__(self,  title, author):
        
        self.title = title
        self.author = author
        self.availaible =True

    def __str__(self):
        return f"{self.title} by {self.author}{' (Available)' if self.availaible else '(Borrowed)'}" 
        
class library:
    def __init__(self):
        self.books =[]

    def add_book(self, book):
        self.books.append(book)
        return f"Added book :{book.title} by {book.author} to the library."
    
    def borrow_book(self,title):
        for book in self.books:
            if book.title == title and book.availaible:
                book.availaible = False
                return f" Borrowed book :{title}."
        return f"Book {title} is not availaible for borrowing."

    def display_availaible_books(self):
        availaible_books = [book for book in self.books if book.availaible]
        if availaible_books :
            return "\n".join(str(book) for book in availaible_books)
        else:
            return "No books availaible in the library "
        

library1 = library()

book1 = Book("The girl in the glass case","devashish parsana")
book2 = Book("400 Days ","Chetan Bhagat")


print(library1.add_book(book1))
print(library1.add_book(book2))

print(library1.display_availaible_books())

print(library1.borrow_book("400 Days"))
print(library1.display_availaible_books())

print(library1.borrow_book("400 Days"))
print(library1.display_availaible_books())

#5.
class product:
    def __init__(self, product_id , name , quantity):
        self.product_id   = product_id
        self.name  = name 
        self.quantity = quantity

    def __str__(self):
        return f"Product Id : {self.product_id}, Name : {self.name }, Quantity:{self.quantity}"
    
class Inventorysystem:
    def __init__(self):
        self.product = []

    def add_product(self,product):
        self.product.append(product)
        return f"Added product : {product.name} to the inventory."
    
    def update_product_quantity(self,product_id,new_quantity):
        for product in self.product:
            if product.product_id == product_id:
                product.quantity == new_quantity
                return f"Updated quantity  for product {product_id} to {new_quantity}"
        return f"Product {product_id} not found in the inventory."
    
    def display_availaible_products(self):
        if self.product:
            return "\n".join(str(product) for product in self.product)
        else:
            return "No products in the inventory."
        

inventory_system = Inventorysystem()
product1 = product(1,"Laptop",10)
product2 = product(2,"Smartphone",20)
product3 = product(3,"Headphones",15)

print(inventory_system.add_product(product1))
print(inventory_system.add_product(product2))
print(inventory_system.add_product(product3))

print(inventory_system.display_availaible_products())

print(inventory_system.update_product_quantity(2,25))
print(inventory_system.update_product_quantity(4,5))

print(inventory_system.display_availaible_products())

class shape:
    def __init__(self ,length,width = None,height=None):
        self.length =length 
        self.width = width
        self.height = height 

    def calculate_area(self):
        if self.width is None and self.height is None:
            
            return f"Area : Not applicable for 1D shape (length: {self.length})"
        elif self.height is None:

            return f"Area : {self.length *self .width}"
        else:
            return f"Surface Area : {2* (self.length * self.width + self.length *self.height + self.width * self.height)}"
        
    def calculate_perimeter(self):
        if self.width is None and self.height is None:
            f"Perimeter :Not applicablel for 1D shape (length: {self.length})"
        elif self.height is None:
            return f"Perimeter :{2* (self.length+ self.width )}"
        else:
            return f"Volume :{self.length* self.width *self.height}"
        
line = shape(5)
rectangle = shape(4,8)
cuboid= shape(3,5,2)

print(line.calculate_area())
print(line.calculate_perimeter())

print(rectangle.calculate_area())
print(rectangle.calculate_perimeter)

print(cuboid.calculate_area())
print(cuboid.calculate_perimeter())

class Student():
    def __init__(self,student_id,name):
        self.student_id = student_id
        self.name = name 
        self.grade =[]

    def add_grade(self,grade):
        self.grade.append(grade)
    
    def calculate_average(self):
        if not self.grade:
            return "No grades availaible for calculations "
        else:
            average_grade = sum(self.grade)/len(self.grade)
            return f" Average grade for {self.name} (Id:{self.student_id}): {average_grade}"
    
    def student_details (self):
        return f" Student id : {self.student_id}\nStudent name: {self.name}"
    

student1 = Student(1,"Rohny")
student2 = Student(2,"Deep")

student1.add_grade(50)
student1.add_grade(40)
student1.add_grade(60)

student2.add_grade(70)
student2.add_grade(60)
student2.add_grade(20)

print(student1.student_details())
print(student1.calculate_average())

print(student2.student_details())
print(student2.calculate_average())


#8.
class email:
    def __init__(self,sender ,recipient,subject):
        self.sender = sender 
        self.recipient = recipient
        self.subject = subject 
        self.sent = False

    def send_email(self):
        
        self.sent = True 
        return f"Email is sent from {self.sender } to {self.recipient} .subject {self.subject}"
    
    def display_email_details(self):
        status ="sent" if self.sent else "draft"
        return f"Sender :{self.sender} ,Recipient ;{self.recipient} and Subject :{self.subject}"
    

email1 = email("devi@gmail.com","kumkum@gmail.com","Invitation")
email2 = email("shubh@gmail.com","manu@gmail.com","Project updation")

print(email1.display_email_details())
print(email1.send_email())
print(email.display_email_details)

print(email2.display_email_details())

#9.
class social_media_profile:
    def __init__(self,username):
        self.username =username
        self.posts = []

    def add_post(self,content):
        self.posts.append(content)
        return f"  post added to {self.username} your profile  "
    
    def display_post(self):
        if  not self.posts:
            return f"{self.username} profile has no post."
        else:
            return f"{self.username}'s posts:\n" + "\n".join(f"{index +1}.{post}" for index ,post in enumerate(self.posts))
    

    def search_posts(self,keyword):
        matching_posts =[post for  post in self.posts if keyword.lower() in post.lower()]
        if not matching_posts:
            return f"No post containig '{keyword}' found in{self.username}'s profile"
        else:
            return f"Matching posts in {self.username}'s profile:\n" + "\n".join(f"{index +1} .{post}" for index, post in enumerate(matching_posts))
        

profile = social_media_profile("riyadubey")

print(profile.add_post("just had a great day at disneyland!"))
print(profile.add_post("Working on a new project.Exciting times!"))
print(profile.add_post("Enjoying me time "))

print(profile.display_post())

search_keyword ="Project"
print(profile.search_posts(search_keyword))

#10.

class ToDolist: 
    def __init__(self):
        self.tasks =[]

    def add_task(self,task,due_date =None):
        new_task ={"task" : task,"due_date": due_date,"completed": False}
        self.tasks.append(new_task)
        return f"Task'{task}' added to the to-do list"
    def mark_task_as_completed(self,task_index):
        if 0<= task_index < len(self.tasks):
            self.tasks[task_index]["completed"] = True
            return f"Task '{self.tasks[task_index]['task']}' markes as completed."
        else:
            return "Invalid task index ."
        
    def display_pending_tasks(self):
        pending_tasks = [task for task in self.tasks if not task["completed"]]
        if not pending_tasks:
            return "No pending tasks in the to-do  list ."
        else:
            return "\n".join(f"Task: {task['task']} ,Due date []:{task['due_date']}"  for task in pending_tasks)
        
to_do_list = ToDolist()

print(to_do_list.add_task("Complete project"))
print(to_do_list.add_task("Buy groceries","2023-01-31"))
print(to_do_list.add_task("Read a chapter"))


print(to_do_list.display_pending_tasks())
print(to_do_list.mark_task_as_completed(0))
print(to_do_list.display_pending_tasks())


                          
```
