```python
##Basic If-Else Statements:

 

#1. Write a Python program to check if a given number is positive or negative.

def check_positive_negative(number):
    if number > 0:
        return "Positive"
    elif number < 0:
        return "Negative"
    else:
        return "Zero"

# Example usage:
user_number = float(input("Enter a number: "))

result = check_positive_negative(user_number)
print(f"The given number is {result}.")

#2. Create a program that determines if a person is eligible to vote based on their age.

def check_voting_eligibility(age):
    if age >= 18:
        return "You are eligible to vote."
    else:
        return "You are not eligible to vote yet."

# Example usage:
user_age = int(input("Enter your age: "))

result = check_voting_eligibility(user_age)
print(result)

#3. Develop a program to find the maximum of two numbers using if-else statements.

def find_maximum(num1, num2):
    if num1 > num2:
        return num1
    else:
        return num2

# Example usage:
number1 = float(input("Enter the first number: "))
number2 = float(input("Enter the second number: "))

maximum_number = find_maximum(number1, number2)
print(f"The maximum of {number1} and {number2} is: {maximum_number}")

#4. Write a Python script to classify a given year as a leap year or not.

def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

# Example usage:
user_year = int(input("Enter a year: "))

if is_leap_year(user_year):
    print(f"{user_year} is a leap year.")
else:
    print(f"{user_year} is not a leap year.")

#5. Create a program that checks whether a character is a vowel or a consonant.

def check_vowel_consonant(character):
    vowels = "aeiouAEIOU"

    if character.isalpha() and len(character) == 1:
        if character in vowels:
            return f"{character} is a vowel."
        else:
            return f"{character} is a consonant."
    else:
        return "Invalid input. Please enter a single alphabet character."

# Example usage:
user_character = input("Enter a character: ")

result = check_vowel_consonant(user_character)
print(result)

#6. Implement a program to determine whether a given number is even or odd.

def check_even_odd(number):
    if number % 2 == 0:
        return f"{number} is even."
    else:
        return f"{number} is odd."

# Example usage:
user_number = int(input("Enter a number: "))

result = check_even_odd(user_number)
print(result)

#7. Write a Python function to calculate the absolute value of a number without using the `abs()` function.

def absolute_value(number):
    if number < 0:
        return -number
    else:
        return number

# Example usage:
user_number = float(input("Enter a number: "))

result = absolute_value(user_number)
print(f"The absolute value of {user_number} is: {result}")

#8. Develop a program that determines the largest of three given numbers using if-else statements.

def find_largest(num1, num2, num3):
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3

# Example usage:
number1 = float(input("Enter the first number: "))
number2 = float(input("Enter the second number: "))
number3 = float(input("Enter the third number: "))

largest_number = find_largest(number1, number2, number3)
print(f"The largest number among {number1}, {number2}, and {number3} is: {largest_number}")

#9. Create a program that checks if a given string is a palindrome.

def is_palindrome(input_string):
    cleaned_string = ''.join(char.lower() for char in input_string if char.isalnum())
    return cleaned_string == cleaned_string[::-1]

# Example usage:
user_string = input("Enter a string: ")

if is_palindrome(user_string):
    print(f"{user_string} is a palindrome.")
else:
    print(f"{user_string} is not a palindrome.")

#10. Write a Python program to calculate the grade based on a student's score

def calculate_grade(score):
    if 90 <= score <= 100:
        return 'A'
    elif 80 <= score < 90:
        return 'B'
    elif 70 <= score < 80:
        return 'C'
    elif 60 <= score < 70:
        return 'D'
    elif 0 <= score < 60:
        return 'F'
    else:
        return 'Invalid score. Please enter a score between 0 and 100.'

# Example usage:
try:
    user_score = float(input("Enter the student's score: "))
    result = calculate_grade(user_score)
    print(f"The student's grade is: {result}")
except ValueError:
    print("Invalid input. Please enter a numeric score.")


##Nested If-Else Statements:

 

#11. Write a program to find the largest among three numbers using nested if-else statements.

# Function to find the largest among three numbers
def find_largest(num1, num2, num3):
    if num1 >= num2:
        if num1 >= num3:
            largest = num1
        else:
            largest = num3
    else:
        if num2 >= num3:
            largest = num2
        else:
            largest = num3

    return largest

# Taking input from the user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))

# Calling the function and displaying the result
result = find_largest(num1, num2, num3)
print(f"The largest number among {num1}, {num2}, and {num3} is: {result}")

#12. Implement a program to determine if a triangle is equilateral, isosceles, or scalene.

def triangle_type(side1, side2, side3):
    if side1 == side2 == side3:
        return "Equilateral"
    elif side1 == side2 or side1 == side3 or side2 == side3:
        return "Isosceles"
    else:
        return "Scalene"

def main():
    # Taking input from the user for the sides of the triangle
    side1 = float(input("Enter the length of the first side: "))
    side2 = float(input("Enter the length of the second side: "))
    side3 = float(input("Enter the length of the third side: "))

    # Checking the type of the triangle and displaying the result
    result = triangle_type(side1, side2, side3)
    print(f"The triangle with sides {side1}, {side2}, and {side3} is a {result} triangle.")

# Calling the main function
main()

#13. Develop a program that checks if a year is a leap year and also if it is a century year.

def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

def is_century_year(year):
    if year % 100 == 0:
        return True
    else:
        return False

def main():
    # Taking input from the user
    year = int(input("Enter a year: "))

    # Checking if it's a leap year and/or a century year
    leap_year = is_leap_year(year)
    century_year = is_century_year(year)

    # Displaying the result
    if leap_year and century_year:
        print(f"{year} is a leap year and a century year.")
    elif leap_year:
        print(f"{year} is a leap year but not a century year.")
    elif century_year:
        print(f"{year} is not a leap year but is a century year.")
    else:
        print(f"{year} is neither a leap year nor a century year.")

# Calling the main function
main()

#14. Write a Python script to determine if a number is positive, negative, or zero.

# Function to determine if a number is positive, negative, or zero
def check_number(n):
    if n > 0:
        return "Positive"
    elif n < 0:
        return "Negative"
    else:
        return "Zero"

# Taking input from the user
number = float(input("Enter a number: "))

# Calling the function and displaying the result
result = check_number(number)
print(f"The number {number} is {result}.")

#15. Create a program to check if a person is a teenager (between 13 and 19 years old).

def is_teenager(age):
    if 13 <= age <= 19:
        return True
    else:
        return False

# Taking input from the user
age = int(input("Enter the person's age: "))

# Checking if the person is a teenager and displaying the result
if is_teenager(age):
    print("The person is a teenager.")
else:
    print("The person is not a teenager.")

#16. Develop a program that determines the type of angle based on its measure (acute, obtuse, or right).

def angle_type(measure):
    if measure < 90:
        return "Acute"
    elif measure == 90:
        return "Right"
    elif measure > 90 and measure < 180:
        return "Obtuse"
    else:
        return "Invalid"

# Taking input from the user
angle_measure = float(input("Enter the measure of the angle: "))

# Checking the type of the angle and displaying the result
result = angle_type(angle_measure)

if result == "Invalid":
    print("Invalid angle measure. Please enter a valid angle measure.")
else:
    print(f"The angle with measure {angle_measure} degrees is a {result} angle.")

#17. Write a Python program to calculate the roots of a quadratic equation.

import math

def calculate_roots(a, b, c):
    # Calculate the discriminant
    discriminant = b**2 - 4*a*c

    # Check the nature of the roots
    if discriminant > 0:
        # Two distinct real roots
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return "Two distinct real roots", root1, root2
    elif discriminant == 0:
        # One real root (double root)
        root = -b / (2*a)
        return "One real root (double root)", root
    else:
        # Complex roots
        real_part = -b / (2*a)
        imaginary_part = math.sqrt(abs(discriminant)) / (2*a)
        root1 = f"{real_part} + {imaginary_part}i"
        root2 = f"{real_part} - {imaginary_part}i"
        return "Complex roots", root1, root2

# Taking input from the user for coefficients
a = float(input("Enter the coefficient a: "))
b = float(input("Enter the coefficient b: "))
c = float(input("Enter the coefficient c: "))

# Calling the function and displaying the result
result, *roots = calculate_roots(a, b, c)
print(f"The quadratic equation {a}x^2 + {b}x + {c} = 0 has {result}: {roots}")

#18. Implement a program to determine the day of the week based on a user-provided number (1 for Monday, 2 for Tuesday, etc.).

def get_day_of_week(number):
    days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

    if 1 <= number <= 7:
        return days_of_week[number - 1]
    else:
        return "Invalid input. Please enter a number between 1 and 7."

# Taking input from the user
user_input = int(input("Enter a number (1 for Monday, 2 for Tuesday, etc.): "))

# Calling the function and displaying the result
result = get_day_of_week(user_input)
print(f"The day of the week corresponding to {user_input} is: {result}")

#19. Create a program that determines if a year is a leap year and also if it is evenly divisible by 400.

def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

# Taking input from the user
year = int(input("Enter a year: "))

# Checking if it's a leap year and/or divisible by 400
leap_year = is_leap_year(year)

# Displaying the result
if leap_year:
    print(f"{year} is a leap year and is evenly divisible by 400.")
else:
    print(f"{year} is not a leap year or is not evenly divisible by 400.")

#20. Develop a program that checks if a given number is prime or not using nested if-else statements.

def is_prime(number):
    # Check if the number is less than 2
    if number < 2:
        return False
    # Check for factors from 2 to the square root of the number
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    # If no factors are found, the number is prime
    else:
        return True

# Taking input from the user
user_input = int(input("Enter a number: "))

# Calling the function and displaying the result
if is_prime(user_input):
    print(f"{user_input} is a prime number.")
else:
    print(f"{user_input} is not a prime number.")


##Elif Statements:

 

#21. Write a Python program to assign grades based on different ranges of scores using elif statements.

def assign_grade(score):
    if 90 <= score <= 100:
        return 'A'
    elif 80 <= score < 90:
        return 'B'
    elif 70 <= score < 80:
        return 'C'
    elif 60 <= score < 70:
        return 'D'
    elif 0 <= score < 60:
        return 'F'
    else:
        return 'Invalid input. Please enter a score between 0 and 100.'

# Taking input from the user
user_score = float(input("Enter the score: "))

# Calling the function and displaying the result
result = assign_grade(user_score)
print(f"The grade for a score of {user_score} is: {result}")

#22. Implement a program to determine the type of a triangle based on its angles.

def triangle_type(angle1, angle2, angle3):
    # Check if the angles form a valid triangle
    if angle1 + angle2 + angle3 == 180:
        # Check the type of triangle based on angles
        if angle1 == angle2 == angle3:
            return "Equilateral Triangle"
        elif angle1 == 90 or angle2 == 90 or angle3 == 90:
            return "Right-angled Triangle"
        elif angle1 < 90 and angle2 < 90 and angle3 < 90:
            return "Acute-angled Triangle"
        else:
            return "Obtuse-angled Triangle"
    else:
        return "Invalid triangle angles. The sum of angles should be 180 degrees."

# Taking input from the user for angles
angle1 = float(input("Enter the first angle: "))
angle2 = float(input("Enter the second angle: "))
angle3 = float(input("Enter the third angle: "))

# Calling the function and displaying the result
result = triangle_type(angle1, angle2, angle3)
print(f"The triangle with angles {angle1}, {angle2}, and {angle3} is a {result}.")

#23. Develop a program to categorize a given person's BMI into underweight, normal, overweight, or obese using elif statements.

def calculate_bmi(weight, height):
    # Formula for BMI: weight (kg) / (height (m))^2
    bmi = weight / (height ** 2)
    return bmi

def categorize_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal Weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

# Taking input from the user for weight and height
weight = float(input("Enter weight in kilograms: "))
height = float(input("Enter height in meters: "))

# Calculating BMI
bmi_result = calculate_bmi(weight, height)

# Categorizing BMI and displaying the result
category = categorize_bmi(bmi_result)
print(f"Your BMI is {bmi_result:.2f}, and you are categorized as {category}.")

#24. Create a program that determines whether a given number is positive, negative, or zero using elif statements.

def check_number(num):
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"

# Taking input from the user
user_input = float(input("Enter a number: "))

# Calling the function and displaying the result
result = check_number(user_input)
print(f"The number {user_input} is {result}.")

#25. Write a Python script to determine the type of a character (uppercase, lowercase, or special) using elif statements.

def character_type(char):
    if char.isupper():
        return "Uppercase"
    elif char.islower():
        return "Lowercase"
    elif char.isnumeric() or not char.isalnum():
        return "Special character"
    else:
        return "Unknown"

# Taking input from the user
user_char = input("Enter a character: ")

# Checking the type of the character and displaying the result
result = character_type(user_char)

print(f"The character '{user_char}' is {result}.")

#26. Implement a program to calculate the discounted price based on different purchase amounts using elif statements.

def calculate_discounted_price(original_price, purchase_amount):
    # Applying discounts based on purchase amount
    if purchase_amount >= 1000:
        discount_rate = 0.1
    elif purchase_amount >= 500:
        discount_rate = 0.05
    else:
        discount_rate = 0

    # Calculating discounted price
    discounted_price = original_price - (original_price * discount_rate)
    return discounted_price

# Taking input from the user
original_price = float(input("Enter the original price: "))
purchase_amount = float(input("Enter the purchase amount: "))

# Calling the function and displaying the result
discounted_price = calculate_discounted_price(original_price, purchase_amount)
print(f"The discounted price for a purchase amount of {purchase_amount} is: {discounted_price:.2f}")

#27. Develop a program to calculate the electricity bill based on different consumption slabs using elif statements.

def calculate_electricity_bill(units):
    if units <= 50:
        rate = 0.50
    elif units <= 150:
        rate = 0.75
    elif units <= 250:
        rate = 1.20
    else:
        rate = 1.50

    # Calculating the electricity bill
    bill_amount = units * rate
    return bill_amount

# Taking input from the user
consumed_units = float(input("Enter the number of consumed units: "))

# Calling the function and displaying the result
electricity_bill = calculate_electricity_bill(consumed_units)
print(f"The electricity bill for {consumed_units} units is: ${electricity_bill:.2f}")

#28. Create a program to determine the type of quadrilateral based on its angles and sides using elif statements.

def determine_quadrilateral_type(side1, side2, side3, side4, angle1, angle2, angle3, angle4):
    # Check if the quadrilateral is valid
    if (angle1 + angle2 + angle3 + angle4 == 360) and (side1 + side2 + side3 + side4 > max(side1, side2, side3, side4) * 2):
        # Check the type of quadrilateral based on angles and sides
        if angle1 == angle2 == angle3 == angle4 == 90:
            return "Rectangle"
        elif angle1 == angle3 and angle2 == angle4 and side1 == side3 and side2 == side4:
            return "Rhombus"
        elif side1 == side2 == side3 == side4 and angle1 == angle2 == angle3 == angle4:
            return "Square"
        elif angle1 == angle3 and angle2 == angle4:
            return "Parallelogram"
        else:
            return "Quadrilateral"
    else:
        return "Invalid quadrilateral"

# Taking input from the user for sides and angles
side1 = float(input("Enter the length of side 1: "))
side2 = float(input("Enter the length of side 2: "))
side3 = float(input("Enter the length of side 3: "))
side4 = float(input("Enter the length of side 4: "))
angle1 = float(input("Enter the measure of angle 1: "))
angle2 = float(input("Enter the measure of angle 2: "))
angle3 = float(input("Enter the measure of angle 3: "))
angle4 = float(input("Enter the measure of angle 4: "))

# Calling the function and displaying the result
result = determine_quadrilateral_type(side1, side2, side3, side4, angle1, angle2, angle3, angle4)
print(f"The given quadrilateral is a {result}.")

#29. Write a Python script to determine the season based on a user-provided month using elif statements.


def determine_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    elif month in [9, 10, 11]:
        return "Fall"
    else:
        return "Invalid month. Please enter a month between 1 and 12."

# Taking input from the user
user_month = int(input("Enter the month (1-12): "))

# Calling the function and displaying the result
result = determine_season(user_month)
print(f"The season for month {user_month} is: {result}")

#30. Implement a program to determine the type of a year (leap or common) and month (30 or 31 days) using elif statements.


def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

def determine_days_in_month(month, year):
    # Check if the month is valid (between 1 and 12)
    if 1 <= month <= 12:
        # Check for months with 30 or 31 days
        if month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        elif month == 2:
            # February has 29 days in a leap year, 28 otherwise
            if is_leap_year(year):
                return 29
            else:
                return 28
        else:
            return 30
    else:
        return "Invalid month. Please enter a month between 1 and 12."

# Taking input from the user
user_year = int(input("Enter the year: "))
user_month = int(input("Enter the month (1-12): "))

# Calling the functions and displaying the result
leap_year_result = "leap" if is_leap_year(user_year) else "common"
days_in_month_result = determine_days_in_month(user_month, user_year)

print(f"The year {user_year} is a {leap_year_result} year.")
print(f"The month {user_month} has {days_in_month_result} days.")




##Basic Level:

 

#1. Write a Python program that checks if a given number is positive, negative, or zero.

# Taking input from the user
number = float(input("Enter a number: "))

# Checking if the number is positive, negative, or zero
if number > 0:
    print("The number is positive.")
elif number < 0:
    print("The number is negative.")
else:
    print("The number is zero.")

#2. Create a program to determine if a person is eligible to vote based on their age.

# Taking input from the user
age = int(input("Enter your age: "))

# Checking if the person is eligible to vote
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote yet. Please wait until you turn 18.")

#3. Write a program to find the maximum of two given numbers using conditional statements.

# Taking input from the user
number1 = float(input("Enter the first number: "))
number2 = float(input("Enter the second number: "))

# Finding the maximum of two numbers
if number1 > number2:
    maximum = number1
else:
    maximum = number2

# Displaying the result
print(f"The maximum of {number1} and {number2} is: {maximum}")

#4. Develop a program that calculates the grade of a student based on their exam score.

# Taking input from the user
exam_score = float(input("Enter the exam score: "))

# Calculating the grade based on the exam score
if 90 <= exam_score <= 100:
    grade = 'A'
elif 80 <= exam_score < 90:
    grade = 'B'
elif 70 <= exam_score < 80:
    grade = 'C'
elif 60 <= exam_score < 70:
    grade = 'D'
else:
    grade = 'F'

# Displaying the result
print(f"The student's grade for an exam score of {exam_score} is: {grade}")

#5. Create a program that checks if a year is a leap year or not.

# Taking input from the user
year = int(input("Enter a year: "))

# Checking if it's a leap year
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")

#6. Write a program to classify a triangle based on its sides' lengths.

def classify_triangle(side1, side2, side3):
    if side1 == side2 == side3:
        return "Equilateral Triangle"
    elif side1 == side2 or side1 == side3 or side2 == side3:
        return "Isosceles Triangle"
    else:
        return "Scalene Triangle"

# Taking input from the user
side1 = float(input("Enter the length of side 1: "))
side2 = float(input("Enter the length of side 2: "))
side3 = float(input("Enter the length of side 3: "))

# Calling the function and displaying the result
result = classify_triangle(side1, side2, side3)
print(f"The triangle with sides {side1}, {side2}, and {side3} is a {result}.")

#7. Build a program that determines the largest of three given numbers.

# Taking input from the user
number1 = float(input("Enter the first number: "))
number2 = float(input("Enter the second number: "))
number3 = float(input("Enter the third number: "))

# Determining the largest of three numbers
largest_number = max(number1, number2, number3)

# Displaying the result
print(f"The largest of {number1}, {number2}, and {number3} is: {largest_number}")

#8. Develop a program that checks whether a character is a vowel or a consonant.

# Taking input from the user
char = input("Enter a character: ")

# Checking if the character is a vowel or a consonant
if char.lower() in ['a', 'e', 'i', 'o', 'u']:
    print(f"The character '{char}' is a vowel.")
elif char.isalpha():
    print(f"The character '{char}' is a consonant.")
else:
    print(f"The input '{char}' is not a valid alphabetical character.")

#9. Create a program to calculate the total cost of a shopping cart based on discounts.

def calculate_total_cost(item_prices):
    # Calculate the total cost
    total_cost = sum(item_prices)

    # Apply a discount based on the total cost
    if total_cost >= 100:
        discount = 0.1  # 10% discount for total cost >= 100
    else:
        discount = 0

    # Calculate the discounted price
    discounted_price = total_cost - (total_cost * discount)

    return discounted_price

# Taking input from the user for item prices
num_items = int(input("Enter the number of items in the shopping cart: "))
item_prices = []

for i in range(num_items):
    price = float(input(f"Enter the price of item {i + 1}: "))
    item_prices.append(price)

# Calling the function and displaying the result
total_cost = calculate_total_cost(item_prices)
print(f"The total cost of the shopping cart is: ${total_cost:.2f}")

#10. Write a program that checks if a given number is even or odd.

 # Taking input from the user
number = int(input("Enter a number: "))

# Checking if the number is even or odd
if number % 2 == 0:
    print(f"The number {number} is even.")
else:
    print(f"The number {number} is odd.")


##Intermediate Level:

 

#11. Write a program that calculates the roots of a quadratic equation .

import math

def calculate_roots(a, b, c):
    # Calculate the discriminant
    discriminant = b**2 - 4*a*c

    # Check the nature of the roots
    if discriminant > 0:
        # Two distinct real roots
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return "Two distinct real roots", root1, root2
    elif discriminant == 0:
        # One real root (double root)
        root = -b / (2*a)
        return "One real root (double root)", root
    else:
        # Complex roots
        real_part = -b / (2*a)
        imaginary_part = math.sqrt(abs(discriminant)) / (2*a)
        root1 = f"{real_part} + {imaginary_part}i"
        root2 = f"{real_part} - {imaginary_part}i"
        return "Complex roots", root1, root2

# Taking input from the user for coefficients
a = float(input("Enter the coefficient a: "))
b = float(input("Enter the coefficient b: "))
c = float(input("Enter the coefficient c: "))

# Calling the function and displaying the result
result, *roots = calculate_roots(a, b, c)
print(f"The quadratic equation {a}x^2 + {b}x + {c} = 0 has {result}: {roots}")

#12. Create a program that determines the day of the week based on the day number (1-7).

def determine_day_of_week(day_number):
    days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

    # Check if the day number is valid
    if 1 <= day_number <= 7:
        return days_of_week[day_number - 1]
    else:
        return "Invalid day number. Please enter a number between 1 and 7."

# Taking input from the user
user_input = int(input("Enter a number (1 for Monday, 2 for Tuesday, etc.): "))

# Calling the function and displaying the result
result = determine_day_of_week(user_input)
print(f"The day of the week corresponding to {user_input} is: {result}")

#13. Develop a program that calculates the factorial of a given number using recursion.

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Taking input from the user
user_input = int(input("Enter a number to calculate its factorial: "))

# Calling the function and displaying the result
result = factorial(user_input)
print(f"The factorial of {user_input} is: {result}")

#14. Write a program to find the largest among three numbers without using the `max()` function.

# Taking input from the user
number1 = float(input("Enter the first number: "))
number2 = float(input("Enter the second number: "))
number3 = float(input("Enter the third number: "))

# Finding the largest of three numbers without using max()
largest_number = number1

if number2 > largest_number:
    largest_number = number2

if number3 > largest_number:
    largest_number = number3

# Displaying the result
print(f"The largest of {number1}, {number2}, and {number3} is: {largest_number}")

#15. Create a program that simulates a basic ATM transaction menu.

def display_balance(balance):
    print(f"Your current balance is: ${balance:.2f}")

def deposit(balance, amount):
    balance += amount
    print(f"Deposit successful. Your new balance is: ${balance:.2f}")
    return balance

def withdraw(balance, amount):
    if amount > balance:
        print("Insufficient funds. Withdrawal unsuccessful.")
    else:
        balance -= amount
        print(f"Withdrawal successful. Your new balance is: ${balance:.2f}")
    return balance

def main():
    # Initial balance
    current_balance = 1000.00

    while True:
        print("\nATM Menu:")
        print("1. Display Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            display_balance(current_balance)
        elif choice == '2':
            deposit_amount = float(input("Enter the deposit amount: $"))
            current_balance = deposit(current_balance, deposit_amount)
        elif choice == '3':
            withdraw_amount = float(input("Enter the withdrawal amount: $"))
            current_balance = withdraw(current_balance, withdraw_amount)
        elif choice == '4':
            print("Exiting ATM. Have a nice day!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

if __name__ == "__main__":
    main()

#16. Build a program that checks if a given string is a palindrome or not.

def is_palindrome(input_str):
    # Removing spaces and converting to lowercase for case-insensitive comparison
    cleaned_str = ''.join(input_str.split()).lower()

    # Comparing the original and reversed strings
    return cleaned_str == cleaned_str[::-1]

# Taking input from the user
user_input = input("Enter a string: ")

# Checking if the string is a palindrome and displaying the result
if is_palindrome(user_input):
    print("The entered string is a palindrome.")
else:
    print("The entered string is not a palindrome.")

#17. Write a program that calculates the average of a list of numbers, excluding the smallest and largest values.

def calculate_average(numbers):
    # Checking if the list has at least 3 elements
    if len(numbers) < 3:
        return "Insufficient elements to calculate average."

    # Removing the smallest and largest values
    numbers.remove(min(numbers))
    numbers.remove(max(numbers))

    # Calculating the average of the remaining values
    average = sum(numbers) / len(numbers)
    return average

# Taking input from the user for a list of numbers
try:
    input_numbers = list(map(float, input("Enter a list of numbers separated by spaces: ").split()))
except ValueError:
    print("Invalid input. Please enter numeric values.")
    exit()

# Calling the function and displaying the result
result = calculate_average(input_numbers)
if isinstance(result, str):
    print(result)
else:
    print(f"The average of the list (excluding the smallest and largest values) is: {result:.2f}")

#18. Develop a program that converts a given temperature from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Taking input from the user
celsius_temperature = float(input("Enter the temperature in Celsius: "))

# Calling the function and displaying the result
fahrenheit_temperature = celsius_to_fahrenheit(celsius_temperature)
print(f"{celsius_temperature} degrees Celsius is equal to {fahrenheit_temperature:.2f} degrees Fahrenheit")

#19. Create a program that simulates a basic calculator for addition, subtraction, multiplication, and division.

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: Division by zero."
    return x / y

# Displaying the calculator menu
print("Basic Calculator:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

# Taking input from the user
choice = input("Enter your choice (1-4): ")

if choice in ['1', '2', '3', '4']:
    # Taking input for operands
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))

    # Performing the selected operation
    if choice == '1':
        result = add(num1, num2)
        operation = "Addition"
    elif choice == '2':
        result = subtract(num1, num2)
        operation = "Subtraction"
    elif choice == '3':
        result = multiply(num1, num2)
        operation = "Multiplication"
    else:
        result = divide(num1, num2)
        operation = "Division"

    # Displaying the result
    print(f"{operation} result: {result:.2f}")
else:
    print("Invalid choice. Please enter a number between 1 and 4.")

#20. Write a program that determines the roots of a cubic equation using the Cardano formula.

import cmath  # Complex math module for handling complex numbers

def solve_cubic_equation(a, b, c, d):
    # Check for a special case when 'a' is zero (quadratic equation)
    if a == 0:
        print("This is not a cubic equation.")
        return

    # Calculate discriminant and other necessary values
    delta0 = c**2 - 3*b*d
    delta1 = 2*c**3 - 9*b*c*d + 27*a*d**2
    C = ((delta1 + cmath.sqrt(delta1**2 - 4*delta0**3)) / 2)**(1/3)

    # Calculate the roots using Cardano formula
    root1 = -1 / (3*a) * (c + C + delta0 / C)
    root2 = -1 / (3*a) * (c + cmath.exp(2j * cmath.pi / 3) * C + delta0 / (C * cmath.exp(2j * cmath.pi / 3)))
    root3 = -1 / (3*a) * (c + cmath.exp(4j * cmath.pi / 3) * C + delta0 / (C * cmath.exp(4j * cmath.pi / 3)))

    return root1, root2, root3

# Taking input from the user for coefficients
try:
    a = float(input("Enter the coefficient a: "))
    b = float(input("Enter the coefficient b: "))
    c = float(input("Enter the coefficient c: "))
    d = float(input("Enter the coefficient d: "))
except ValueError:
    print("Invalid input. Please enter numeric values.")
    exit()

# Calling the function and displaying the result
roots = solve_cubic_equation(a, b, c, d)
print(f"The roots of the cubic equation are: {roots}")

 

##Advanced Level:

 

#21. Create a program that calculates the income tax based on the user's income and tax brackets.

def calculate_income_tax(income):
    # Define tax brackets and corresponding tax rates
    brackets = [0, 10000, 40000, 80000, float('inf')]
    rates = [0.10, 0.20, 0.30, 0.40]

    # Initialize variables
    total_tax = 0
    previous_bracket = 0

    # Calculate tax for each bracket
    for i in range(1, len(brackets)):
        if income > brackets[i]:
            taxable_amount = brackets[i] - previous_bracket
        else:
            taxable_amount = income - previous_bracket

        total_tax += taxable_amount * rates[i - 1]
        previous_bracket = brackets[i]

    return total_tax

# Taking input from the user
try:
    user_income = float(input("Enter your annual income: $"))
except ValueError:
    print("Invalid input. Please enter a numeric value for income.")
    exit()

# Calling the function and displaying the result
income_tax = calculate_income_tax(user_income)
print(f"Your income tax for the year is: ${income_tax:.2f}")

#22. Write a program that simulates a rock-paper-scissors game against the computer.

import random

def get_user_choice():
    while True:
        user_choice = input("Enter your choice (rock, paper, or scissors): ").lower()
        if user_choice in ["rock", "paper", "scissors"]:
            return user_choice
        else:
            print("Invalid choice. Please enter 'rock', 'paper', or 'scissors'.")

def get_computer_choice():
    return random.choice(["rock", "paper", "scissors"])

def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!"
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "paper" and computer_choice == "rock") or
        (user_choice == "scissors" and computer_choice == "paper")
    ):
        return "You win!"
    else:
        return "Computer wins!"

# Main game loop
while True:
    # Get user and computer choices
    user_choice = get_user_choice()
    computer_choice = get_computer_choice()

    # Display choices
    print(f"\nYour choice: {user_choice}")
    print(f"Computer's choice: {computer_choice}")

    # Determine and display the winner
    result = determine_winner(user_choice, computer_choice)
    print(result)

    # Ask the user if they want to play again
    play_again = input("\nDo you want to play again? (yes/no): ").lower()
    if play_again != 'yes':
        print("Thanks for playing. Goodbye!")
        break

#23. Develop a program that generates a random password based on user preferences (length, complexity).

import random
import string

def generate_password(length, include_lowercase, include_uppercase, include_digits, include_symbols):
    characters = ""
    
    if include_lowercase:
        characters += string.ascii_lowercase
    if include_uppercase:
        characters += string.ascii_uppercase
    if include_digits:
        characters += string.digits
    if include_symbols:
        characters += string.punctuation

    if not characters:
        print("Please include at least one character type.")
        return None

    password = ''.join(random.choice(characters) for _ in range(length))
    return password

# Get user preferences
try:
    length = int(input("Enter the desired password length: "))
except ValueError:
    print("Invalid input. Please enter a numeric value for length.")
    exit()

include_lowercase = input("Include lowercase letters? (yes/no): ").lower() == 'yes'
include_uppercase = input("Include uppercase letters? (yes/no): ").lower() == 'yes'
include_digits = input("Include digits? (yes/no): ").lower() == 'yes'
include_symbols = input("Include symbols? (yes/no): ").lower() == 'yes'

# Generate and display the password
password = generate_password(length, include_lowercase, include_uppercase, include_digits, include_symbols)

if password:
    print(f"Generated Password: {password}")

#24. Create a program that implements a simple text-based adventure game with branching scenarios.
import time

def introduction():
    print("Welcome to the Text Adventure Game!")
    time.sleep(1)
    print("You find yourself in a dark, mysterious forest.")
    time.sleep(1)
    print("Your goal is to navigate through the forest and reach the treasure.")
    time.sleep(1)

def make_choice(prompt, options):
    while True:
        print(prompt)
        for i, option in enumerate(options, start=1):
            print(f"{i}. {option}")

        try:
            choice = int(input("Enter your choice (1, 2, 3, ...): "))
            if 1 <= choice <= len(options):
                return choice
            else:
                print("Invalid choice. Please enter a number within the given options.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def forest_scenario():
    print("You venture deeper into the forest.")
    time.sleep(1)
    print("Ahead, you see a path that splits into three directions.")

    choices = ["Take the left path.", "Take the middle path.", "Take the right path."]
    choice = make_choice("Which path will you choose?", choices)

    if choice == 1:
        print("You encounter a river blocking your way.")
        time.sleep(1)
        print("Do you want to swim across or find another way?")
        choices = ["Swim across.", "Find another way."]
        choice = make_choice("What will you do?", choices)
        if choice == 1:
            print("You swim across the river and continue your journey.")
        else:
            print("You find a bridge upstream and cross safely.")
    elif choice == 2:
        print("You find a hidden shortcut that leads you closer to the treasure.")
    else:
        print("The path becomes dark and dangerous. You decide to turn back.")

def treasure_scenario():
    print("Congratulations! You have reached the treasure.")
    time.sleep(1)
    print("You open the chest and find it filled with gold and precious jewels.")
    time.sleep(1)
    print("You've successfully completed the adventure!")

def main():
    introduction()

    # Start of the adventure
    forest_scenario()

    # Continue to the treasure scenario
    treasure_scenario()

if __name__ == "__main__":
    main()

#25. Build a program that solves a linear equation  for x, considering different cases.

def solve_linear_equation(a, b):
    if a == 0:
        if b == 0:
            return "Infinite solutions (identity)"
        else:
            return "No solution (contradiction)"
    else:
        x = -b / a
        return f"The solution to the linear equation {a}x + {b} = 0 is x = {x:.2f}"

# Taking input from the user for coefficients
try:
    a = float(input("Enter the coefficient a: "))
    b = float(input("Enter the coefficient b: "))
except ValueError:
    print("Invalid input. Please enter numeric values.")
    exit()

# Calling the function and displaying the result
result = solve_linear_equation(a, b)
print(result)

#26. Write a program that simulates a basic quiz game with multiple-choice questions and scoring.

import time

def ask_question(question, options, correct_option):
    print(question)
    for i, option in enumerate(options, start=1):
        print(f"{i}. {option}")

    while True:
        try:
            user_answer = int(input("Enter the number of your answer: "))
            if 1 <= user_answer <= len(options):
                break
            else:
                print("Invalid choice. Please enter a number within the given options.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    return user_answer == correct_option

def run_quiz(questions):
    score = 0

    for question, options, correct_option in questions:
        if ask_question(question, options, correct_option):
            print("Correct!\n")
            score += 1
        else:
            print(f"Wrong! The correct answer was {correct_option}. \n")

        # Pause for a moment before the next question
        time.sleep(1)

    return score

def main():
    print("Welcome to the Quiz Game!\n")
    time.sleep(1)

    quiz_questions = [
        ("What is the capital of France?", ["Paris", "Berlin", "Madrid"], 1),
        ("Which planet is known as the Red Planet?", ["Mars", "Venus", "Jupiter"], 1),
        ("What is the largest mammal in the world?", ["Elephant", "Blue Whale", "Giraffe"], 2),
    ]

    user_score = run_quiz(quiz_questions)

    print(f"Quiz completed! Your score is: {user_score}/{len(quiz_questions)}")

if __name__ == "__main__":
    main()

#27. Develop a program that determines whether a given year is a prime number or not.

def is_prime(year):
    if year < 2:
        return False
    for i in range(2, int(year**0.5) + 1):
        if year % i == 0:
            return False
    return True

# Taking input from the user
try:
    user_year = int(input("Enter a year: "))
except ValueError:
    print("Invalid input. Please enter a numeric value for the year.")
    exit()

# Calling the function and displaying the result
if is_prime(user_year):
    print(f"{user_year} is a prime year.")
else:
    print(f"{user_year} is not a prime year.")

#28. Create a program that sorts three numbers in ascending order using conditional statements.

# Taking input from the user
try:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    num3 = float(input("Enter the third number: "))
except ValueError:
    print("Invalid input. Please enter numeric values.")
    exit()

# Sorting the numbers in ascending order
if num1 <= num2 <= num3:
    sorted_numbers = (num1, num2, num3)
elif num1 <= num3 <= num2:
    sorted_numbers = (num1, num3, num2)
elif num2 <= num1 <= num3:
    sorted_numbers = (num2, num1, num3)
elif num2 <= num3 <= num1:
    sorted_numbers = (num2, num3, num1)
elif num3 <= num1 <= num2:
    sorted_numbers = (num3, num1, num2)
else:
    sorted_numbers = (num3, num2, num1)

# Displaying the sorted numbers
print("Sorted numbers in ascending order:", sorted_numbers)

#29. Build a program that determines the roots of a quartic equation using numerical methods.

def quartic_equation(x, coefficients):
    return sum(a * x**i for i, a in enumerate(coefficients))

def derivative_quartic(x, coefficients):
    return sum(i * a * x**(i - 1) for i, a in enumerate(coefficients) if i > 0)

def newton_method(initial_guess, coefficients, tolerance=1e-6, max_iterations=100):
    x = initial_guess

    for _ in range(max_iterations):
        f_x = quartic_equation(x, coefficients)
        f_prime_x = derivative_quartic(x, coefficients)

        if abs(f_x) < tolerance:
            return x

        x = x - f_x / f_prime_x

    print("Newton's method did not converge.")
    return None

def main():
    print("Quartic Equation Solver using Newton's Method\n")

    try:
        coefficients = [float(input(f"Enter the coefficient for x^{i}: ")) for i in range(4, -1, -1)]
    except ValueError:
        print("Invalid input. Please enter numeric values for coefficients.")
        exit()

    try:
        initial_guess = float(input("Enter an initial guess for the root: "))
    except ValueError:
        print("Invalid input. Please enter a numeric value for the initial guess.")
        exit()

    root = newton_method(initial_guess, coefficients)

    if root is not None:
        print(f"\nThe approximate root is: {root:.6f}")

if __name__ == "__main__":
    main()

#30.Write a program that calculates the BMI (Body Mass Index) and provides health recommendations based on the user's input.

def calculate_bmi(weight, height):
    # BMI formula: weight (kg) / (height (m))^2
    bmi = weight / (height ** 2)
    return bmi

def interpret_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal (Healthy) Weight"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

def main():
    print("BMI Calculator and Health Recommendations\n")

    try:
        weight = float(input("Enter your weight in kilograms: "))
        height = float(input("Enter your height in meters: "))
    except ValueError:
        print("Invalid input. Please enter numeric values for weight and height.")
        exit()

    # Calculate BMI
    bmi = calculate_bmi(weight, height)

    # Interpret BMI and provide health recommendations
    bmi_category = interpret_bmi(bmi)

    print(f"\nYour BMI is: {bmi:.2f}")
    print(f"BMI Category: {bmi_category}")

    if bmi_category == "Underweight":
        print("Recommendation: Consider consulting with a healthcare professional for advice.")
    elif bmi_category == "Normal (Healthy) Weight":
        print("Recommendation: Maintain a balanced diet and regular exercise for overall health.")
    elif bmi_category == "Overweight":
        print("Recommendation: Consider adopting a healthier lifestyle with a focus on diet and exercise.")
    else:
        print("Recommendation: Consult with a healthcare professional to discuss potential health risks and management.")

if __name__ == "__main__":
    main()

 

##Challenge Level:

 

#31. Create a program that validates a password based on complexity rules (length, characters, etc.).

import re

def is_valid_password(password):
    # Check if the password meets the length requirement (at least 8 characters)
    if len(password) < 8:
        return False

    # Check if the password contains at least one uppercase letter
    if not any(char.isupper() for char in password):
        return False

    # Check if the password contains at least one lowercase letter
    if not any(char.islower() for char in password):
        return False

    # Check if the password contains at least one digit
    if not any(char.isdigit() for char in password):
        return False

    # Check if the password contains at least one special character
    if not re.search("[!@#$%^&*(),.?\":{}|<>]", password):
        return False

    # All checks passed, the password is valid
    return True

def main():
    print("Password Validator\n")

    user_password = input("Enter your password: ")

    if is_valid_password(user_password):
        print("Password is valid. Good job!")
    else:
        print("Password is not valid. Please follow the complexity rules.")

if __name__ == "__main__":
    main()

#32. Develop a program that performs matrix addition and subtraction based on user input.

def matrix_addition(matrix1, matrix2):
    result = [[0 for _ in range(len(matrix1[0]))] for _ in range(len(matrix1))]
    
    for i in range(len(matrix1)):
        for j in range(len(matrix1[0])):
            result[i][j] = matrix1[i][j] + matrix2[i][j]
    
    return result

def matrix_subtraction(matrix1, matrix2):
    result = [[0 for _ in range(len(matrix1[0]))] for _ in range(len(matrix1))]
    
    for i in range(len(matrix1)):
        for j in range(len(matrix1[0])):
            result[i][j] = matrix1[i][j] - matrix2[i][j]
    
    return result

def input_matrix():
    rows = int(input("Enter the number of rows: "))
    cols = int(input("Enter the number of columns: "))
    
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = float(input(f"Enter the element at position ({i + 1}, {j + 1}): "))
            row.append(element)
        matrix.append(row)
    
    return matrix

def display_matrix(matrix, title):
    print(f"\n{title}:")
    for row in matrix:
        print(row)

def main():
    print("Matrix Addition and Subtraction\n")

    print("Enter details for the first matrix:")
    matrix1 = input_matrix()

    print("\nEnter details for the second matrix:")
    matrix2 = input_matrix()

    # Check if matrices can be added or subtracted
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        print("\nMatrices must have the same dimensions for addition or subtraction.")
    else:
        # Perform matrix addition and subtraction
        result_addition = matrix_addition(matrix1, matrix2)
        result_subtraction = matrix_subtraction(matrix1, matrix2)

        # Display results
        display_matrix(matrix1, "Matrix 1")
        display_matrix(matrix2, "Matrix 2")
        display_matrix(result_addition, "Matrix Addition Result")
        display_matrix(result_subtraction, "Matrix Subtraction Result")

if __name__ == "__main__":
    main()

#33. Write a program that calculates the greatest common divisor (GCD) of two numbers using the Euclidean algorithm.

def euclidean_algorithm(a, b):
    while b:
        a, b = b, a % b
    return a

def main():
    print("Greatest Common Divisor (GCD) Calculator using Euclidean Algorithm\n")

    try:
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
    except ValueError:
        print("Invalid input. Please enter integer values for the numbers.")
        exit()

    if num1 < 0 or num2 < 0:
        print("Please enter non-negative integers.")
        exit()

    gcd = euclidean_algorithm(num1, num2)

    print(f"\nThe GCD of {num1} and {num2} is: {gcd}")

if __name__ == "__main__":
    main()

#34. Build a program that performs matrix multiplication using nested loops and conditional statements.

def matrix_multiplication(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        print("Error: The number of columns in the first matrix must be equal to the number of rows in the second matrix.")
        return None

    result = [[0 for _ in range(len(matrix2[0]))] for _ in range(len(matrix1))]

    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    return result

def input_matrix(rows, cols):
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = float(input(f"Enter the element at position ({i + 1}, {j + 1}): "))
            row.append(element)
        matrix.append(row)

    return matrix

def display_matrix(matrix, title):
    print(f"\n{title}:")
    for row in matrix:
        print(row)

def main():
    print("Matrix Multiplication using Nested Loops and Conditional Statements\n")

    try:
        rows1 = int(input("Enter the number of rows for the first matrix: "))
        cols1 = int(input("Enter the number of columns for the first matrix: "))
        rows2 = int(input("Enter the number of rows for the second matrix: "))
        cols2 = int(input("Enter the number of columns for the second matrix: "))
    except ValueError:
        print("Invalid input. Please enter integer values for matrix dimensions.")
        exit()

    if cols1 != rows2:
        print("Error: The number of columns in the first matrix must be equal to the number of rows in the second matrix.")
        exit()

    print("\nEnter details for the first matrix:")
    matrix1 = input_matrix(rows1, cols1)

    print("\nEnter details for the second matrix:")
    matrix2 = input_matrix(rows2, cols2)

    result = matrix_multiplication(matrix1, matrix2)

    if result:
        # Display original matrices and result
        display_matrix(matrix1, "Matrix 1")
        display_matrix(matrix2, "Matrix 2")
        display_matrix(result, "Matrix Multiplication Result")

if __name__ == "__main__":
    main()

#35. Create a program that simulates a basic text-based tic-tac-toe game against the computer.

import random

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 17)

def check_winner(board, player):
    # Check rows and columns
    for i in range(3):
        if all(board[i][j] == player for j in range(3)) or all(board[j][i] == player for j in range(3)):
            return True

    # Check diagonals
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True

    return False

def is_board_full(board):
    return all(board[i][j] != " " for i in range(3) for j in range(3))

def player_move(board):
    while True:
        try:
            row = int(input("Enter the row (1, 2, or 3): ")) - 1
            col = int(input("Enter the column (1, 2, or 3): ")) - 1

            if 0 <= row < 3 and 0 <= col < 3 and board[row][col] == " ":
                return row, col
            else:
                print("Invalid move. Please choose an empty cell.")
        except ValueError:
            print("Invalid input. Please enter numeric values for the row and column.")

def computer_move(board):
    empty_cells = [(i, j) for i in range(3) for j in range(3) if board[i][j] == " "]
    return random.choice(empty_cells)

def main():
    print("Welcome to Tic-Tac-Toe!\n")
    board = [[" " for _ in range(3)] for _ in range(3)]
    player_symbol = "X"
    computer_symbol = "O"

    while True:
        print_board(board)

        # Player's move
        player_row, player_col = player_move(board)
        board[player_row][player_col] = player_symbol

        # Check if the player wins
        if check_winner(board, player_symbol):
            print_board(board)
            print("Congratulations! You win!")
            break

        # Check if the board is full (tie)
        if is_board_full(board):
            print_board(board)
            print("It's a tie!")
            break

        # Computer's move
        print("\nComputer's turn:")
        computer_row, computer_col = computer_move(board)
        board[computer_row][computer_col] = computer_symbol

        # Check if the computer wins
        if check_winner(board, computer_symbol):
            print_board(board)
            print("Computer wins! Better luck next time.")
            break

if __name__ == "__main__":
    main()

#36. Write a program that generates Fibonacci numbers up to a specified term using iterative methods.

def generate_fibonacci(n):
    fibonacci_sequence = [0, 1]

    for i in range(2, n):
        next_term = fibonacci_sequence[-1] + fibonacci_sequence[-2]
        fibonacci_sequence.append(next_term)

    return fibonacci_sequence[:n]

def main():
    print("Fibonacci Number Generator (Iterative)\n")

    try:
        terms = int(input("Enter the number of Fibonacci terms to generate: "))
        if terms <= 0:
            print("Please enter a positive integer for the number of terms.")
            return
    except ValueError:
        print("Invalid input. Please enter a positive integer.")
        return

    fibonacci_numbers = generate_fibonacci(terms)

    print(f"\nThe first {terms} Fibonacci numbers are:")
    print(fibonacci_numbers)

if __name__ == "__main__":
    main()

#37. Develop a program that calculates the nth term of the Fibonacci sequence using memoization.

def fibonacci_memo(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci_memo(n - 1, memo) + fibonacci_memo(n - 2, memo)
    return memo[n]

def main():
    print("Fibonacci Number Calculator using Memoization\n")

    try:
        n = int(input("Enter the value of n for the nth Fibonacci term: "))
        if n < 0:
            print("Please enter a non-negative integer for n.")
            return
    except ValueError:
        print("Invalid input. Please enter a non-negative integer.")
        return

    result = fibonacci_memo(n)

    print(f"\nThe {n}th Fibonacci term is: {result}")

if __name__ == "__main__":
    main()

#38. Create a program that generates a calendar for a given month and year using conditional statements.

import calendar

def generate_calendar(year, month):
    cal = calendar.monthcalendar(year, month)
    month_name = calendar.month_name[month]
    print(f"\n{month_name} {year}\n")
    print("Mo Tu We Th Fr Sa Su")

    for week in cal:
        week_str = ""
        for day in week:
            if day == 0:
                week_str += "   "
            else:
                week_str += f"{day:2} "
        print(week_str.strip())

def main():
    print("Calendar Generator\n")

    try:
        year = int(input("Enter the year: "))
        month = int(input("Enter the month (1-12): "))
        if not (1 <= month <= 12):
            print("Invalid month. Please enter a number between 1 and 12.")
            return
    except ValueError:
        print("Invalid input. Please enter numeric values for the year and month.")
        return

    generate_calendar(year, month)

if __name__ == "__main__":
    main()

#39. Build a program that simulates a basic text-based blackjack game against the computer.

import random

def deal_card():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    return random.choice(cards)

def calculate_score(cards):
    if sum(cards) == 21 and len(cards) == 2:
        return 0  # Blackjack
    if 11 in cards and sum(cards) > 21:
        cards.remove(11)
        cards.append(1)
    return sum(cards)

def compare(user_score, computer_score):
    if user_score > 21 and computer_score > 21:
        return "You went over. You lose!"

    if user_score == computer_score:
        return "It's a draw!"
    elif computer_score == 0:
        return "Computer has a Blackjack. You lose!"
    elif user_score == 0:
        return "You have a Blackjack! You win!"
    elif user_score > 21:
        return "You went over. You lose!"
    elif computer_score > 21:
        return "Computer went over. You win!"
    elif user_score > computer_score:
        return "You win!"
    else:
        return "You lose!"

def play_game():
    user_cards = []
    computer_cards = []
    is_game_over = False

    for _ in range(2):
        user_cards.append(deal_card())
        computer_cards.append(deal_card())

    while not is_game_over:
        user_score = calculate_score(user_cards)
        computer_score = calculate_score(computer_cards)

        print(f"  Your cards: {user_cards}, current score: {user_score}")
        print(f"  Computer's first card: {computer_cards[0]}")

        if user_score == 0 or computer_score == 0 or user_score > 21:
            is_game_over = True
        else:
            should_continue = input("Type 'y' to get another card, or 'n' to pass: ")
            if should_continue == 'y':
                user_cards.append(deal_card())
            else:
                is_game_over = True

    while computer_score != 0 and computer_score < 17:
        computer_cards.append(deal_card())
        computer_score = calculate_score(computer_cards)

    print(f"  Your final hand: {user_cards}, final score: {user_score}")
    print(f"  Computer's final hand: {computer_cards}, final score: {computer_score}")
    print(compare(user_score, computer_score))

def main():
    print("Welcome to Blackjack!\n")
    play_game()

if __name__ == "__main__":
    main()

#40.Write a program that generates the prime factors of a given number using trial division.

def prime_factors(n):
    factors = []
    divisor = 2
    
    while n > 1:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 1
    
    return factors

# Example usage:
number = int(input("Enter a number: "))
result = prime_factors(number)

print(f"Prime factors of {number}: {result}")
```
